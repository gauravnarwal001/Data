```python
q1 = df['total_bill'].quantile(0.25)
q3 = df['total_bill'].quantile(0.75)
```

```python
print(q1)
print(q3)
```

```python
iqr = q3 - q1
iqr
```

```python
upper_fence = (1.5 * iqr) + q3
lower_fence = (1.5 * iqr) - q1
```

```python
print(upper_fence)
print(lower_fence)
```

```python
new_df = df[(df['total_bill'] >= lower_fence) & (df['total_bill'] <= upper_fence)]
new_df
```

```python
plt.boxplot(new_df['total_bill'])
plt.show()
```

```python
def remove_outliers(column,df = None):
    
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    
    iqr = q3 - q1
    
    upper_fence = (1.5 * iqr) + q3
    lower_fence = (1.5 * iqr) - q1
    

    
    if df is not None:
        if len((column < upper_fence) & (column >= lower_fence)) != len(df):
            raise ValueError(f'Length of df argument is not match with columns length')
        try:
            return df[(column < upper_fence) & (column >= lower_fence)]
        except:
            raise TypeError(f"{df} is not DataFrame")
            
    else:
    
        return (column < upper_fence) & (column >= lower_fence)
```

```python
remove_outliers(df['total_bill'],df = None)
```
